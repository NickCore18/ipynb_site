- P(Disease)=0.02

Test sensitivity (probability of a positive result given disease)
- P(Positive∣Disease)=0.95

False positive rate (probability of a positive result given no disease)
- P(Positive∣NoDisease)=0.05

Probability of a positive result in the population
- P(Positive)=0.06


---

**Computation using the Formula**

Substituting values into Bayes' Theorem Formula:

- P(Disease∣Positive)= P(Positive∣Disease) × P(Disease) / P(Positive)

​
Insert the values:

- P(Disease∣Positive)= 0.95 × 0.02 / 0.06

- 0.95 × 0.02 = 0.019

- P(Disease∣Positive)= 0.019 / 0.06

- **P(Disease∣Positive) = 0.3167**(approximately)

# Part 4. Visualize a Probability Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from matplotlib.colors import LinearSegmentedColormap

n = 5
p = 0.5

x = np.arange(0, n + 1)

probabilities = binom.pmf(x, n, p)

colors = LinearSegmentedColormap.from_list("red_to_purple", ["#FF0000", "#800080"], N=len(x))

plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color=colors(np.linspace(0, 1, len(x))), alpha=0.7)  # Apply gradient colors
plt.xticks(x)
plt.xlabel('Number of Heads', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.title('Binomial Distribution: Probability of Getting Heads in 5 Coin Flips', fontsize=16)
plt.ylim(0, max(probabilities) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i in range(len(x)):
    plt.text(x[i], probabilities[i], f'{probabilities[i]:.2f}', ha='center', va='bottom')

plt.show()
```
