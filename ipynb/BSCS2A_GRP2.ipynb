**1.** Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

**2.** Loading Dataset

```python
em_data = pd.read_csv('Email_Exchanges.csv',encoding="latin-1")
iv_data = pd.read_csv('Inventory_Records.csv',encoding="latin-1")
sc_data = pd.read_csv('Security_Camera_Logs.csv',encoding="latin-1")
vl_data = pd.read_csv('Visitor_Logs.csv',encoding="latin-1")
```

**3.** Data Loading and Initial Analysis

```python
email_statistics = em_data.describe()
inventory_statistics = iv_data.describe()
security_statistics = sc_data.describe()
visitor_statistics = vl_data.describe()
```

**4.** Data Manipulation

```python
email_data = em_data.groupby('Date').sum()
inv_data = iv_data.groupby('Movement Date').sum()
sec_data = sc_data.groupby('Timestamp').sum()

vis_data = vl_data.groupby('Date').sum()
vl_data = vl_data.sort_index(ascending=True)
```

**5.** Data Visualization



---


*Email_Exchanges.csv*

```python
plt.figure(figsize=(20, 12),facecolor='#bff4ff')

plt.plot(em_data['Date'],em_data['From'],label='From',marker='o')
plt.plot(em_data['Date'],em_data['To'],label='To',marker='o')

plt.vlines(em_data['Date'],ymin=em_data[['From','To']].min(axis=1),ymax=em_data[['From','To']].max(axis=1),linestyle='dotted')
plt.xlabel('Dates',fontweight='bold')
plt.ylabel('Emails',fontweight='bold')

plt.title('Email Exchanges',fontsize=14,fontweight='extra bold')
plt.text(0.99,0.99,'Graph By Group 2 BSCS-2A',horizontalalignment='right',verticalalignment='top',transform=plt.gca().transAxes,fontsize=11,fontweight='light')
plt.grid(axis='y')
plt.legend(loc='lower right')

plt.show()
```

*Inventory_Records.csv*

```python
fig, ax1 = plt.subplots(figsize=(20, 12), facecolor='#bff4ff')

ax1.plot(iv_data['Movement Date'],iv_data['Authorized Personnel'])
ax1.set_xlabel('Movement Date',fontweight='bold')
ax1.set_ylabel('Authorized Personnel',fontweight='bold')

ax2 = ax1.twinx()

ax2.plot(iv_data['Movement Date'],iv_data['Last Known Location'],color='orange')
ax2.set_ylabel('Last Known Location',fontweight='bold')

ax3 = ax1.twiny()

ax3.plot(iv_data['Description'],iv_data['Authorized Personnel'],marker='o',label='Description')

plt.title('Inventory Records', fontsize=14, fontweight='extra bold')
plt.text(0.99, 0.99, 'Graph By Group 2 BSCS-2A',horizontalalignment='right',verticalalignment='top',transform=plt.gca().transAxes,fontsize=11,fontweight='light')
plt.legend(loc='lower right')

plt.show()
```

*Security_Camera_Logs.csv*

```python
fig, ax1 = plt.subplots(figsize=(20, 12), facecolor='#bff4ff')

ax1.plot(sc_data['Location'], sc_data['Timestamp'],label='Timestamp')
ax1.set_xlabel('Location')
ax1.set_ylabel('Timestamp')

ax2 = ax1.twinx()

ax2.plot(sc_data['Camera ID'], label='Camera ID', color='orange')
ax2.set_ylabel('Camera ID')

ax3 = ax1.twiny()

ax3.plot(label='Description')
ax3.set_ylabel('Description')

plt.title('Security Camera Logs', fontsize=14, fontweight='extra bold')
plt.text(0.99, 0.99, 'Graph By Group 2 BSCS-2A',horizontalalignment='right',verticalalignment='top',transform=plt.gca().transAxes,fontsize=11,fontweight='light')

plt.show()
```

*Visitor_Logs.csv*

```python
plt.figure(figsize=(20, 12),facecolor='#bff4ff')
plt.plot(vl_data['Exit Time'],vl_data['Date'],label='Entry Time',marker='o')
plt.plot(vl_data['Entry Time'],vl_data['Date'],label='Exit Time',marker='o')
# plt.vlines(vl_data['Date'],ymin=vl_data[['Entry Time','Exit Time']].min(axis=1),ymax=vl_data[['Entry Time','Exit Time']].max(axis=1),linestyle='dotted',color='black')
plt.xlabel('Dates',fontweight='bold')
plt.ylabel('Emails',fontweight='bold')
plt.title('Email Exchanges',fontsize=14,fontweight='extra bold')
plt.text(0.99,0.99,'Graph By Group 2 BSCS-2A',horizontalalignment='right',verticalalignment='top',transform=plt.gca().transAxes,fontsize=11,fontweight='light')
plt.grid(axis='both')
plt.legend(loc='lower right')
plt.show()
```
